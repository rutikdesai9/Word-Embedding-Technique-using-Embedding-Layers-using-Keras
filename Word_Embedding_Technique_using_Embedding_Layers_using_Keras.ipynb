```
### Importing Libraries
import tensorflow as tf
import keras
#### one_hot libray converts words in sentence to some specific index
from tensorflow.keras.preprocessing.text import one_hot
```

```
#### we provide sentence
sent=['the glass of milk',
      'i am good boy',
      'the glass of juice',
      'the cup of tea',
      'I am good developer',
      'understand meaning of words',
      'your ideos are good',
      'its her choice',]
```

```
sent
```

```
#### if we give this size the index of words will be between 0 to vocab size
### Vocabulory Size provided
vocab_size=10000
```

```
#### one_hot library used
onehot_repr=[one_hot(words,vocab_size)for words in sent]
```

```
onehot_repr
```

# **Word Embedding Representation**

```
##### Importing Libraries
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential #### for model
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```
import numpy as np
```

```
#### pad sequence library used
sent_length=8 ### by these number we get array of 8 rows. it starts filling from back side. remaining columns will fill with 0
embedded_docs=pad_sequences(sequences=onehot_repr,padding='pre',maxlen=sent_length)
```

```
embedded_docs
```

```
### features for embedding layer
dim=10
```

```
model=Sequential()
```

```
#### Embedding Layer starts
model.add(Embedding(vocab_size,dim,input_length=sent_length))
```

```
### Compiling model with Adam optimizer and accuracy is mean-squared-error
model.compile('adam','mse')
```

```
model.summary()
```

```
### prediction of model
model.predict(embedded_docs)
```
